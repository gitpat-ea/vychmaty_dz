#### 1.3 Подсчитайте след матрицы и количество ненулевых элементов на ее диагонали.

```python
M = np.array([[1, 2, 3],
              [4, 0, 6],
              [7, 8, 9]])

trace = np.trace(M)
non_zero = np.count_nonzero(np.diag(M))

print(trace)
print(non_zero)
```

#### 3.2 Определить оптимальный шаг численного дифференцирования $h_{\text{opt}}$ при использовании для вычисления производной приближенной формулы

#### $$u^{\prime}(x) \approx \frac{u(x-2 h)-8 u(x-h)+8 u(x+h)-u(x+2 h)}{12 h},$$

#### Какой порядок аппроксимации имеет эта формула? Известно, что, $\left|u^{(5)}(t)\right| \leq M_5$, а значения функций вычисляются с точностью $\Delta u$.

$u(x-h) = u(x) - u'(x) \cdot h + u''(x) \cdot \frac{h^2}{2} - u'''(x)\cdot \frac{h^3}{6} + u''''(x)\cdot \frac{h^4}{24} - u'''''(x)\cdot \frac{h^5}{120}+ O(h^6)$  
$u(x+h) = u(x) + u'(x) \cdot h + u''(x) \cdot \frac{h^2}{2} + u'''(x)\cdot \frac{h^3}{6} + u''''(x)\cdot \frac{h^4}{24} + u'''''(x)\cdot \frac{h^5}{120}+ O(h^6)$  
$u(x-2h) = u(x) - u'(x) \cdot 2h + u''(x) \cdot 2h^2 - u'''(x)\cdot \frac{4h^3}{3} + u''''(x)\cdot \frac{2h^4}{3} - u'''''(x)\cdot \frac{4h^5}{15} + O(h^6)$  
$u(x+2h) = u(x) + u'(x) \cdot 2h + u''(x) \cdot 2h^2 + u'''(x)\cdot \frac{4h^3}{3} + u''''(x)\cdot \frac{2h^4}{3} + u'''''(x)\cdot \frac{4h^5}{15} + O(h^6)$  

$I = u(x+h) - u(x-h) = 2u'(x) \cdot h + u'''(x)\cdot \frac{h^3}{3} + u'''''(x)\cdot \frac{h^5}{60}+ O(h^6)$  
$II = u(x-2h) - u(x+2h) = -4u'(x) \cdot h - \frac{8}{3}u'''(x)\cdot h^3 - \frac{8}{15}u'''''(x)\cdot h^5  + O(h^6)$ 

$\dfrac{II + 8I}{12h} = \dfrac{12u'(x) \cdot h - \frac{8}{3}u'''(x)\cdot h^3 - \frac{8}{15}u'''''(x)\cdot h^5  + 16u'(x) \cdot h + 8u'''(x)\cdot \frac{h^3}{3} + 8u'''''(x)\cdot \frac{h^5}{60}+ O(h^6)}{12h} = \dfrac{12u'(x) \cdot h - \frac{2}{5}u'''''(x)\cdot h^5 + O(h^6)}{12h}$

$u^{\prime}(x, h) = u'(x) - \frac{1}{30}u'''''(x)\cdot h^4 + O(h^5) = u'(x) + O(h^4)$

$\varepsilon_{method} = \frac{1}{30}u'''''(x)\cdot h^4 + O(h^5) \leq \frac{1}{30}\cdot M_5 h^4$

С другой стороны, при численном расчёте на компьютере возникает ошибка округления. Посчитаем её...

$\varepsilon_{round} = \dfrac{18h\Delta u + (u(x-2 h)-8 u(x-h)+8 u(x+h)-u(x+2 h))\Delta h}{12 h^2} \approx \dfrac{3\Delta u}{2 h}$

$\varepsilon_{total} = \varepsilon_{method} + \varepsilon_{round} = \frac{1}{30}\cdot M_5 h^4 + \frac{3\Delta u}{2 h}$  
$\varepsilon^{\prime}_{total} = \frac{2}{15}\cdot M_5 h^3 - \frac{3\Delta u}{2h^2} = 0$  
  
Тогда оптимальное значение шага:  
$h_{opt} = \sqrt[5]{\frac{45\Delta u}{4 M_5}}$

#### 4.1 С каким числом верных знаков (или относительной погрешностью) должен быть известен свободный член в уравнении $x^2-2 x+0.999993751=0$, чтобы корни имели четыре верных знака?

Пусть $с = 0.999993751$ известно с погрешностью $\Delta c$  
$D = b^2 - 4ac = 4 - 4c$  
Тогда $x_{1,2} = \dfrac{2 \pm 2\sqrt{1-c}}{2} = 1 \pm \sqrt{1-c}$  
Для того чтобы корни имели четыре верных знака, ошибка в вычислении корней должна быть меньше $0.00005$. При этом  
$\Delta x = \dfrac{\Delta c}{2\sqrt{1-c}}$  
А значит $\Delta c = 2 \Delta x \sqrt{1-c} = 2 \cdot 0.00005 \cdot \sqrt{1-0.999993751} \approx 2.5\cdot 10^{-7}$  
тогда $\delta c = \dfrac{\Delta c}{c} \approx 2.5 \cdot 10^{-7}$


P.S. здесь и далее $\Delta x$ -- абсолютная погрешность, $\delta x$ -- относительная погрешность.

#### 6.1* (Вигнеровский полкруг). Сгенерируйте случайную матрицу $N \times N$, в которой каждый элемент взят из гауссова нормального распределения с фиксированными параметрами. Сделайте из этой матрицы симметричную, добавив к ней ее транспонированную. Постройте гистограмму собственных значений получившейся матрицы (подберите достаточно большое $N$ и удачное количество бинов). Повторите эксперимент $M$ раз ( $M \sim 50$, здесь цикл разрешается) и отобразите разброс высоты каждого бина гистограммы с помощью креста ошибок.

```python
import numpy as np
import matplotlib.pyplot as plt

matrix_size = 400
num_experiments = 200
mean, std_dev = 0, 1

all_eigenvalues = np.zeros((num_experiments, matrix_size))
for i in range(num_experiments):
    random_matrix = np.random.normal(mean, std_dev, (matrix_size, matrix_size))
    symmetric_matrix = random_matrix + random_matrix.T
    all_eigenvalues[i] = np.linalg.eigvalsh(symmetric_matrix)

avg_eigenvalues = np.mean(all_eigenvalues, axis=0)

errors_diff = np.zeros((num_experiments, matrix_size))
for i in range(num_experiments):
    errors_diff = all_eigenvalues[i] - avg_eigenvalues

max_errors = np.max(errors_diff, axis=0)

num_bins = 25
hist_counts, hist_bins = np.histogram(avg_eigenvalues, bins=num_bins)

bins_to_remove = [0]
updated_bins = np.delete(hist_bins, bins_to_remove)

plt.bar(x=updated_bins, width=5, height=hist_counts, yerr=max_errors, capsize=2, align='edge')
plt.xlabel('Eigenvalue, $\lambda$')
plt.ylabel('Count')
plt.show()
```
